```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i,init='k-means++',random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('The Number of cluster')
plt.ylabel('WCSS')
plt.show();
```

```python
km = KMeans(n_clusters=3,init='k-means++',random_state=42)
y_km = km.fit_predict(x)
```

```python
plt.scatter(x[y_km == 0,0],x[y_km == 0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_km == 1,0],x[y_km == 1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(x[y_km == 2,0],x[y_km == 2,1],s=100,c='green',label='Cluster 3')
#plt.scatter(x[y_km == 3,0],x[y_km == 3,1],s=100,c='cyan',label='Cluster 4')
#plt.scatter(x[y_km == 4,0],x[y_km == 4,1],s=100,c='magenta',label='Cluster 5')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='yellow',label='centroid')


plt.title('Cluster of clustomera')
plt.xlabel('Что то')
plt.ylabel('Что то')
plt.legend()
plt.show()
```

```python
sns.set(font_scale = 1.5)
f, ax = plt.subplots(figsize=(10, 10))
g = sns.scatterplot(x = x[:,0], y = x[:,1], hue = y_km, s = 300, alpha = 0.8)# alpha  прозрачность
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s = 250, marker = '*', c = 'red', label = 'центроиды') # * - форма, например, s  квадратик, 
g.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol = 1)
```

```python
df_for_reg_scal = scal.fit_transform(df_for_reg.drop('new_cases',axis=1))
```

```python
x = df_for_reg_scal.copy()
y = df.new_cases
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

#### LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

Отличный результат

Построим график

```python
pred_train = lr.predict(x_train)

pred_test = lr.predict(x_test)

plt.figure(figsize=(20,5))
plt.scatter(x_train,y_train,label='Data train')
plt.scatter(x_test,y_test,label='Data test')
plt.scatter(x_train,pred_train,label='predict train')
plt.scatter(x_test,pred_test,label='predict test')
plt.legend();
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=10,random_state=0)
```

```python
rfr.fit(x_train,y_train)
```

```python
print("Результат на обучающей выборке = ",rfr.score(x_train,y_train))
print("Результат на тестовой выборке = ",rfr.score(x_test,y_test))
```

```python
pred_train = rfr.predict(x_train)

pred_test = rfr.predict(x_test)

plt.figure(figsize=(20,5))
plt.scatter(x_train,y_train,label='Data train')
plt.scatter(x_test,y_test,label='Data test')
plt.scatter(x_train,pred_train,label='predict train')
plt.scatter(x_test,pred_test,label='predict test')
plt.legend();
```
