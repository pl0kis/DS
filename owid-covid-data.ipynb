###### Посмотрим количество пропусков

```python
print(df.isna().sum()/len(df)*100)
```

Видим большое количество пропусков.
- Удалим столбцы имеющие количество пропусков более 45%
Тк информации они не несут ( из за количества пропусков )

```python
for i in df:
    if df[i].isna().sum()/len(df)*100 > 45:
        del df[i]
```

```python
print(df.isna().sum()/len(df)*100)
```

Отлично. Столбцы с большим количеством пропусков мы удалили.

###### Столбец iso_code ( Код страны )

```python
df.iso_code.unique()
```

Здесь у нас большое количество столбцов.
- Закодируем с помощью LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

Цикл для кодирования большого количества столбцов.

for i in df:
    if df[i].dtype == object:
        df[i] = lb.fit_transform(df[i])

```python
df.iso_code = lb.fit_transform(df.iso_code)
```

###### Столбец continent

```python
df.continent.unique()
```

В столбце "Континент" у нас 7 уникальных значений
- Закодируем с помощью словаря
- При более подробной работе nan можно отнести к континенту смотря на код страны
- Но мы nan заменим на моду

```python
df.continent.value_counts()
```

```python
dictContinent = {'Asia':1, np.nan:3, 'Europe':0, 'Africa':3, 'North America':4, 'South America':5,
       'Oceania':2}
df.continent = df.continent.map(dictContinent)
```

```python
df.continent.unique()
```
